```python
# nested loops
num = int(input("enter a number : "))
if num > 0:
    print("The number is positive")
    if num % 2 == 0:
        print("It's even")
elif num > 1:
    print(num**2)
else:
    print("Wrong number ")
```

```python
##Practical Examples

##Determine if a year  is a leap year using nested condition statement

year = int(input("enter year : "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "It's a leap year")
        else:
            print("This is not a leap year", year)
    else:
        print("This year is leap year", year)
else:
    print("This is a not a leap year", year)
```

```python
# Calculator
# take input from user

a = int(input("Enter number 1:"))
b = int(input("Enter number 2:"))

print("Make a Choice \n1. Addition \n2.Subtraction\n3.Division \n4.Multiplication")
op = int(input("Enter operation you want to perform: "))
if op == 1:
    print("Addition: ", a + b)
elif op == 2:
    print("Subtraction: ", a - b)
elif op == 3:
    print("Division: ", a % b)
elif op == 4:
    print("Multiplication: ", a * b)
else:
    print("Wrong input")
```
