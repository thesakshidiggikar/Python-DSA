```python
# while loop


##it continues to execute until the condition is true:


##temprary variable

count = 0
while count < 5:
    print(count)
    count = count + 1
```

```python
count = 0
while count % 2 == 0:
    print(count)
    print()
    count = count + 1
```

```python
# loop control statements

##break

for i in range(20):
    if i == 6:
        break
    print(i)
```

```python
# continue

##The continues statement skips the current iteration and continues with the next.

for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
##pass
##the pas statement is a null operation; it does nothing.

for i in range(6):
    if i == 3:
        print("the number is,"i)
        pass
    print(i)
```

```python
# nested loop
## a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j: {j}")
```

```python
# examples
# calculate the sum of first n natural numbers
n = 10
sum = 0
count = 1
while count <= n:
    sum = sum + count
    count = count + 1
print("the sum is ", sum)
```

```python
# calculate the sum of 10 natural numbers using for loop
sum = 0
for i in range(11):
    sum = sum + i

print(sum)
```

```python
# example - prime number between 1 and 100

for i in range(1, 101):
    if i > 1:
        for j in range(2, i):
            if i % j == 0:
                break
        else:
            print(i)
```
