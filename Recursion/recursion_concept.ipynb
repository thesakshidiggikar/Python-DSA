## Basic Recursion without Recursion function

```python
def greet():
    print("hello")


greet()
```

## Infinite recursion

```python
def greet():
    print("hello world")
    greet()  # infinite recursion


greet()
```

## Head Recursion

```python
# print hello world 4 times using recursion
count = 0


def greet():
    global count
    if count == 4:
        return
    print("Hello World")
    count += 1
    greet()


greet()
```

## Tail Recursion

```python
count = 0


def greet():
    global count
    if count == 4:
        return
    count += 1
    greet()
    print("Hello World")


greet()
```

## Recursion using parameter

```python
# recursion loop  using recursion parameter and without count variable

#Head Recursion
def greet(x, n):
    if n == 0:
        return
    print(x)
    greet(x, n - 1)


greet("Hello World", 4)
```
