### The Filter Function
##### the filter function constructs an interator from element of an iterable for which a functiion return true.It is used to filter out items from a list(or any other iterable) based on a condition.

```python
def even(num):
    if num % 2 == 0:
        return True


even(20)
```

```python
list_num = [1, 2, 4, 5, 7, 3, 4, 6, 6, 7]
list(filter(even, list_num))
```

```python
##lambda with filter

# find greater then 5

g = list(filter(lambda list_num: list_num > 5, list_num))
```

```python
g
```

```python
# filter with mupltipple contion
# with lambda function
even_and_greater_then_five = list(filter(lambda x: x > 5 and x % 2 == 0, list_num))
even_and_greater_then_five
```

```python
# filter with dict
dict1 = {"name": "A", "age": 30, "city": "England"}
```

```python
dict1 = [
    {"name": "A", "age": 30, "city": "England"},
    {"name": "B", "age": 20, "city": "India"},
    {"name": "C", "age": 28, "city": "USA"},
]


def fun(d):
    return d["age"] > 25


result = list(filter(fun, dict1))
print(result)
```
