```python
my_array = [7, 12, 9, 11]
minVal = my_array[0]

for i in my_array:
    if i < minVal:
        minVal = i

print("Lowest Value : ", minVal)
```

### Implementation Using a For Loop

```python
prevVal1 = 0
prevVal2 = 1

for fib in range(20):
    newFeb = prevVal1 + prevVal2
    print(fib, end=" ")
    prevVal1 = prevVal2
    prevVal2 = newFeb
```
