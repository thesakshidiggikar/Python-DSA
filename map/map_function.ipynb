#this fucntion applies a given fucntion to all the items in an input list(or any other iterable) and return a map object (an iterator). This is particularly useful for transforming data in a list comprehension

```python
# why to use
def sq(num):
    return num**2
```

```python
sq(2)
```

```python
# map
num = [0, 2, 3, 4, 5, 6, 7, 8, 9, 10]
list(map(sq, num))
```

```python
##lambda with map

list(map(lambda num: num**2, num))
```

```python
##map multiple iterable
num1 = [1, 2, 3]
num2 = [4, 5, 6]
add = list(map(lambda num1, num2: num1 + num2, num1, num2))
```

```python
add
```

```python
##map to convert list of string to integer

str_num = ["1", "2", "3", "4", "5"]
lst_num = list(map(int, str_num)) #tyecasting
print(lst_num)
```

```python
type(lst_num)
```

```python
## in built function
word = ["a", "b", "C"]
up = list(map(str.upper, word))
print("Capital words", up)
```
