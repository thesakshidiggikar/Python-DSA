```python
dict1 = {1: "a", 2: "b", 3: "c"}
print(type(dict1))
print(dict1)
```

```python
dict2 = {}
print(dict2)
```

```python
##accessing dictionary element
print(dict1[3])
```

```python
##adding new element
dict1[4] = "Sakhi"
print(dict1)
```

```python
dict1.get(1)
```

```python
# modify the key
dict1[1] = "Apple"
print(dict1)
```

```python
# dictionary methods

keys = dict1.keys()
print("keys: ", keys)
```

```python
values = dict1.values()
print(values)
```

```python
# to get key value we write
dict1.items()
```

```python
# shallow copy
dict2 = dict1
print(dict2)
```

```python
dict1[2] = "Banana"
print(dict2)
```

```python
# so in shallow copy
dict2 = dict1.copy()  # shallow copy
```

```python
# iteration over the dictionaries

##you can use loos to iteratioin
for keys in dict1.keys():
    print(keys)
```

```python
for values in dict1.values():
    print(values)
```

```python
for key in dict1.items():
    print(key)
```
