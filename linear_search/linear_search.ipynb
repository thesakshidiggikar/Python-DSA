```python
# find the 70 as target from lst
def linear_search(lst, target):
    size = len(lst)
    for index in range(0, size):
        if lst[index] == target:
            return index
    return -1


lst = [10, 23, 45, 70, 11]
#
target = 70
result = linear_search(lst, target)
print(f"The answer of linear search is : {result} index")
```
