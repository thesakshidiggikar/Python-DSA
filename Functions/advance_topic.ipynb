```python
# function for temperature conversion
def temp(temp, unit):
    """This function converts temperature between Celsius and Fahrenehit"""

    if unit == "C":
        return temp * 9 / 5 + 32  # formula to convert celcius to faherenheit

    elif unit == "F":
        return (temp - 32) * 5 / 9  ##formula to covert fahrenheit to celsius
```

```python
temp(45, "F")
```

```python
# password strength checker
def pswd_strong(passwd):
    """This function checks if the passwod is strong or not"""
    if len(passwd) < 8:
        return False

    if not any(char.isdigit() for char in passwd):
        return False

    if not any(char.islower() for char in passwd):
        return False
    if not any(char.isupper() for char in passwd):
        return False
    if not any(char in "!@#$%^&*()_ " for char in passwd):
        return False

    else:
        print("The password is super strong")
        return True
```
