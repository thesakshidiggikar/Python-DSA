```python
def evn_or_odd(num):
    """This function finds even or odd"""
    if num % 2 == 0:
        print("This is even")
    else:
        print("This is odd")
```

```python
# call the function
evn_or_odd(10)
```

```python
evn_or_odd(27)
```

```python
# variable legth arguments
# positional and keyword arguments


def prime_number(*args):
    for numbers in args:
        print(numbers)
```

```python
prime_number(1, 2, 3, 4, 5, 6, 7)
```

```python
# keywrod argument
def print_dic(**kwargs):
    ##key value arguments
    for key, value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_dic(name="A", age=48, country="B")
```

```python
def both(*args, **kwargs):
    for val in args:
        print(f"Positional Arguments {val}")

    for key, value in kwargs.items():
        print(f"Keyword Argument--- {key}:{value}")
```

```python
both(1, 2, 3, 4, 5, 6, 7, "S", name="A", age=48, country="B")
```
