```python
def add(a, b):
    return a + b
```

```python
add(1, 3)
```

```python
# using lambda function

add = lambda a, b: a + b
```

```python
type(add)
```

```python
print(add(1, 3))
```

```python
def even(num):
    if num % 2 == 0:
        return True
```

```python
even(2)
```

```python
even1 = lambda num: num % 2 == 0
print(even1(20))
```

```python
def add(a, b, c):
    return a + b + c


add(10, 20, 30)
```

```python
add1 = lambda a, b, c: a + b + c
print(add1(10, 20, 30))
```
